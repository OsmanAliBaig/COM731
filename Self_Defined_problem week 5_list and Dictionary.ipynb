```python
# 1. Initial Creation (Dictionary with 3 modules and grades)
module_grades = {
    "Introduction to AI": 65,
    "PPS": 58,
    "MATHS": 72
}

print("--- Current Semester Grades ---")
for module, grade in module_grades.items():
    print(f"{module}: {grade}%")


# 2. Update Operation (PPS grade improves after a coursework remark)
module_grades["PPS"] = 68
print("\n[Update] PPS mark updated to 68%.")


# 3. Deletion Operation (Dropping MATHS from the tracker)
dropped_module = "MATHS"
if dropped_module in module_grades:
    module_grades.pop(dropped_module)
    print(f"[Delete] Dropped {dropped_module} from this semester.")


# 4. Display the remaining modules
print("\n--- Final Grade Summary ---")
for module, grade in module_grades.items():
    print(f"{module}: {grade}%")
```
