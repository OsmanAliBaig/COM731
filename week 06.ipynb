## Open and Close a csv file

- <font size="3">We use the built-in <span style="color: green;">open()</span> function and the keyword <span style="color: green;">with</span> and <span style="color: green;">as</span> in order to automatically close a file once we are done processing it</font>
- <font size="3">The with statement simplifies exception handling by encapsulating common preparation and cleanup tasks.</font>
- <font size="3"><b>Syntax:</b></font><br>
&emsp;&emsp;&emsp;&emsp;<font size="3"> <span style="color: green;">with open</span>(file_name,mode,encoding='UTF-8') <span style="color: green;">as</span> file_object</font>:<br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;#do something with the file<br>
<div class="alert alert-block alert-info">
<b>Note:</b> open() function creates a file object and associates it with a file on the disk</div><br>
- <font size="3"><b>mode</b> is a character specifying how the file will be opened:</font>
    - <font size="3"><b>read mode ("r")</b>: opens the specified file so contents can be retrieved; default</font>
    - <font size="3"><b>write mode ("w")</b>: opens the specified file so that you can add data, starting at the beginning of the file (creates or overwrites the file)</font>
    - <font size="3"><b>append mode ("a")</b>: opens the specified file so that you can add content to the end of a file that already contains data</font>
- <font size="3"><b>encoding</b> is the process of translating characters into a format that can be stored and transmitted by a computer. In CSV files, encoding determines how the characters in the file are represented. UTF-8 and Latin-1 are two common encoding formats used in CSV files.</font>

```python
#import the libaray
import csv

#define a file name 
file_name = "data/people.csv"

#open the csv file and display if the file is loaded succcessfilly
with open(file_name, mode = "r" , encoding='UTF-8'):
  print(f"{file_name} is opened successfully to read")
```

## Read data from a csv file

- <font size="3">To read a CSV file in Python, we can use the <span style="color: blue;"> csv.reader()</span> function</font>
- <font size="3"><b>Syntax:</b><br>
&emsp;&emsp;&emsp;&emsp;<font size="3"> variable = <span style="color: blue;">csv.reader</span>(file_object)</font>
- <font size="3">A CSV reader allows us to retrieve a row of data one at a time from the file</font>
- <font size="3">When used with a loop, the reader object will return a list of values for each line in the file</font>

### Example: open csv using open() function and read a csv file using csv reader

```python
# import the library
import csv

while True:
    file_name = input("Enter the file name to be opened: ")

    if file_name.lower() == "quit":
        print("Quitting the program!")
        break

    try:
      with open(file_name, mode="r", encoding="UTF-8") as csv_file:
            #print(f"{file_name} is opened successfully loaded")
            print(f"{file_name} is opened successfully to read")
            # creating a reader variable to read the file
            csv_reader = csv.reader(csv_file)
            
            header = next(csv_reader)  # Read the header row
            print(f"Header: {header}")

            for row in csv_reader:
                print(row)  # Print each row of the CSV file  

                break  # Exit the loop after reading the first file
    
          
    except FileNotFoundError as e:
        print(f"{file_name} not found: {e}")
 
```

### Example: reading headings from csv file

```python
# import the library
import csv
#define a list to store the data
data = []

while True:
    file_name = input("Enter the file name to be opened: ")

    if file_name.lower() == "quit":
        print("Quitting the program!")
        break

    try:
      with open(file_name, mode="r", encoding="UTF-8") as csv_file:
            #print(f"{file_name} is opened successfully loaded")
            print(f"{file_name} is opened successfully to read")
            # creating a reader variable to read the file
            csv_reader = csv.reader(csv_file)
            
            header = next(csv_reader)  # Read the header row
            print(f"Header: {header}")

            for row in csv_reader:
                data.append(row)  # Store each row in the data list
                print(row)  # Print each row of the CSV file  

                break  # Exit the loop after reading the first file
    
          
    except FileNotFoundError as e:
        print(f"{file_name} not found: {e}")

#display the header and data stored in the list
print(f"Header: {header}")
print("Data stored in the list:")
for row in data:
    print(row)
 
```

## Reading Data from a CSV File and Storing It in a List
When working with data stored in a CSV file, it's often inefficient to open and read the file every time you need to access the data. Instead, you can read the file once, store the data in a list (or another suitable data structure), and then use that list throughout your program.
- <font size=2>Why Do This?</font>
    - <font size=2>Efficiency: Reading from memory (a list) is much faster than reading from a file repeatedly.</font>
    - <font size=2>
Convenience: You can easily manipulate, filter, or analsze the data once it' loaded.</font>
    - <font size=2>
Cleaner Code: Your code becomes mo organised d and easier to maintai</font>n.

## Access individual element in header and data

```python
# import the library
import csv
#define a list to store the data
data = []

while True:
    file_name = input("Enter the file name to be opened: ")

    if file_name.lower() == "quit":
        print("Quitting the program!")
        break

    try:
      with open(file_name, mode="r", encoding="UTF-8") as csv_file:
            #print(f"{file_name} is opened successfully loaded")
            print(f"{file_name} is opened successfully to read")
            # creating a reader variable to read the file
            csv_reader = csv.reader(csv_file)
            
            header = next(csv_reader)  # Read the header row
           

            for row in csv_reader:
                data.append(row)  # Store each row in the data list
                print(row)  # Print each row of the CSV file  

                break  # Exit the loop after reading the first file
    
          
    except FileNotFoundError as e:
        print(f"{file_name} not found: {e}")

#display the header and data stored in the list
print(f"Header: {header}")
print("Data stored in the list:")
for row in data:
    print(row[0], row[0], row[2])  # Print each row of the data list
 
```

## Write data to a csv file

### write() function

- <font size=2>Python provides a built-in method called <span style="color: blue;">write()</span> that allows you to write text data directly into a file. </font>
- <font size=2><span style="color: blue;">write()</span> adds text to the file exactly as given.</font>
- <font size="2">Write data use commas to separate values.</font>
- <font size=2>It does not automatically add a newline (\n), so you must include it if you want line breaks.</font>
- <font size=2>If the file already exists in 'w' mode, its contents will be overwritten.</font>
- <font size=2>If the file doesn’t exist, Python will create it.</font>

### Exmaple: write() 

```python
import csv
#setting the path to the csv file
#which is stored in the data folder
file_name = "data/movies.csv"

try:
    with open(file_name, mode="w", encoding="UTF-8") as file:
      number = input("how many movies do you want to add in the file?: ")
      #the heading of the csv file is written in the first line of the file
      file.write("Title,Genre,Rating\n")

      #enter the data into the file
      for i in range(int(number)):
          title = input("Enter the title of the movie: ")
          genre = input("Enter the genre of the movie: ")
          rating = input("Enter the rating of the movie: ")

          #write the data into the file
          file.write(f"{title},{genre},{rating}\n")
      #notiy the user that the data is written successfully into the file
      print(f"{number} movies are added successfully into the file {file_name}")

except Exception as e:
    print(f"An error occurred: {e}")

      
      
```

### writer(), writerow() and writerows() from csv module

Python’s csv module provides a simple way to write data into CSV (Comma-Separated Values) files. To do this:
- <font size=2>you use the csv.writer() function to create a writer object that is linked to a file.</font>
- <font size=2>then use either writerow() or writerows() to write data.</font>
    - <font size=2>writerow writes a single row to the CSV file. Each row must be passed as a list (or other iterable).</font>
    - <font size=2>writerows writes multiple rows in one call. You pass a <b>list of lists</b>, a table-like structure—similar to rows and columns in a spreadsheet or CSV file.</font>
- <font size=2>Always open the file with newline='' to avoid extra blank lines on Windows.</font>

### Example: writer(), writerow() and writerows()

```python
# import the library
import csv

file_name = "data/bank.csv"

#open the csv file to write the data into the file

try:
    with open(file_name, mode="w", newline='') as csv_file:
        #create a writer object to write the data into the file
        csv_writer = csv.writer(csv_file)
        #write the header of the csv file
        header = ["Account Number", "Account Holder Name", "Balance"]
        data = [
            ["123456", "John Doe", 1000.50],
            ["789012", "Jane Smith", 2500.75],
            ["345678", "Alice Johnson", 500.00],
            ["901234", "Bob Brown", 1500.25]]
        csv_writer.writerow(header)
        csv_writer.writerows(data)
        print(f"Data is written successfully into the file {file_name}")
except Exception as e:
    print(f"An error occurred: {e}")
```
